```python
#
ndvi_columns_train = [col for col in train.columns if col not in ['ID', 'class']]
ndvi_columns_test = [col for col in test.columns if col != 'ID']


train['ndvi_mean'] = train[ndvi_columns_train].mean(axis=1)
train['ndvi_std'] = train[ndvi_columns_train].std(axis=1)
train['ndvi_min'] = train[ndvi_columns_train].min(axis=1)
train['ndvi_max'] = train[ndvi_columns_train].max(axis=1)
train['ndvi_median'] = train[ndvi_columns_train].median(axis=1)
train['ndvi_range'] = train['ndvi_max'] - train['ndvi_min']

# Same features for test set
test['ndvi_mean'] = test[ndvi_columns_test].mean(axis=1)
test['ndvi_std'] = test[ndvi_columns_test].std(axis=1)
test['ndvi_min'] = test[ndvi_columns_test].min(axis=1)
test['ndvi_max'] = test[ndvi_columns_test].max(axis=1)
test['ndvi_median'] = test[ndvi_columns_test].median(axis=1)
test['ndvi_range'] = test['ndvi_max'] - test['ndvi_min']

# Display to verify
train[['ndvi_mean', 'ndvi_std', 'ndvi_min', 'ndvi_max', 'ndvi_median', 'ndvi_range']].head()
```

```python
train[ndvi_columns_train] = train[ndvi_columns_train].rolling(window=3, axis=1, min_periods=1).mean()
test[ndvi_columns_test] = test[ndvi_columns_test].rolling(window=3, axis=1, min_periods=1).mean()
```
